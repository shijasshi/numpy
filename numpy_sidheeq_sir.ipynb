- 4 is placed at index **4** because it should **come after the last element 3**.
- 6 is also placed at index **4**, as it would **come after the last element 3**.
- In summary, while **the searchsorted() function can still technically operate on an unsorted array, the results may not be meaningful in the context of maintaining order, as the function assumes the array is sorted**.

#Summation Over an Axis

#NumPy GCD greatest common Denominator

##finding GCD(Greatest common denominator)
*  The GDC(Greatest common Denominatot) also knowm as HCF (Highest common factor) is the biggest number that is a common factor of both of the numbers.

```python
#find the HCF of the following to numbers:
import numpy as np
num1=6
num2=9
x=np.gcd(num1,num2)
print(x)
```

#Returns: 3 because that is the highest number both numbers can be divided by (6/3=2 and 9/3-3).

* The np.gcd() function is used to compute the GCD of num1 and num2. The GCD is the largest integer that divides both numbers without leaving a remainder. Q 100%
* In this case, the GCD of 6 and 9 is 3 because 3 is the largest number that divides both 6 and 9.


#Explanation of GCD (Greatest Common Denominator):
* For 6 and 9: The divisors of 6: 1, 2, 3, 6
* The divisors of 9: 1, 3, 9 The common divisors: 1 and 3
* The largest common divisor is 3, which makes the GCD of 6 and 9 equal to 3.

# Finding GCD in Arrays
* To find the Highest Common Factor of all values in an array, you can use the reduce() method.
* The reduce() method will use the ufunc, in this case the gcd() function, on each element, and reduce the array by one dimension.

```python
# Find the GCD for all of the numbers in the following array 100
import numpy as np
arr=np.array([20,8,32,36,16])
x=np.gcd.reduce(arr)
print(x)
```

Returns: 4 because that is the highest number all values can be divided by

* The function np.gcd.reduce(arr) computes the GCD of the entire array by applying the GCD function cumulatively across all elements in the array.
* The reduce function applies the GCD operation iteratively:
    *    It first finds the GCD of the first two numbers (gcd(20, 8) = 4).
    *    Then, it takes this result and applies the GCD with the next number (gcd(4, 32) = 4), and so on.
    *    This process continues until all the numbers in the array are reduced to their greatest common divisor

    

* Detailed GCD Calculation:
    *Finding the GCD of the first two numbers:
    *gcd(20, 8) = 4
        * Finding the GCD of the result with the next number:
    *gcd(4, 32) = 4
        *Next:
    *gcd(4, 36) = 4
        *Final Step:
    *gcd(4, 16) = 4
  *The GCD of all the numbers in the array [20, 8, 32, 36, 16) is 4.

#Create Sets in NumPy

We can use NumPy's unique() method to find unique elements from any array E.g. create a set array, but remember that the set arrays should only be 1-D arrays.

```python
#Convert following array with repeated elements to set:
import numpy as np
arr=np.array([1,1,1,2,3,4,5,5,6,7])
x=np.unique(arr)
print(x)
```

#Finding Union

 To find the unique values of two arrays, use the union1d() method.

```python
#Find union of the following two set arrays:
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([3,4,5,6])
newarr=np.union1d(arr1,arr2)
print(newarr)
```

Note: the intersect1d() method takes an optional argument assumque, which ift to True can speed up computation. It should always be set to True when dealing with sets.

#Finding Difference

 To find only the values in the first set that is NOT present in the seconds set, use the setdiff1d() method.


```python
#Find the difference of the seti from set2:
import numpy as np
set1=np.array([1,2,3,4])
set2=np.array([3,4,5,6])
newarr = np.setdiff1d(set1, set2, assume_unique=True)
print(newarr)
```

Note: the setdiff1d() method takes an optional argument assume_unique, which if set to True can speed up computation. It should always be set to True when dealing with sets.

#Finding Symmetric Difference

 To find only the values that are NOT present in BOTH sets, use the setxor1d() method.

```python
# Find the symmetric difference of the set1 and set2:
import numpy as np
set1=np.array([1,2,3,4])
set2=np.array([3,4,5,6])
newarr=np.setxor1d(set1,set2,assume_unique=True)
print(newarr)
```

Note: the setxor1d() method takes an optional argument assume_unique, which if set to True can speed up computation. It should always be set to True when dealing with sets.
